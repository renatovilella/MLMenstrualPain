```python
from autogluon.tabular import TabularDataset

train_data = TabularDataset("Data\menstrualdata.csv")
```

```python
from autogluon.tabular import TabularPredictor

predictor = TabularPredictor(label="MenstrualPainIntensity").fit(train_data)
```

```python
predictor.leaderboard()
```

```python
new_data = TabularDataset("Data\menstrualdatatocomplete.csv")
predictions = predictor.predict(new_data)
```

```python
predictor.refit_full()
```

```python
new_data["Predicted_MenstrualPainIntensity"] = predictions
```

```python
from autogluon.tabular import TabularDataset
from autogluon.tabular import TabularPredictor
import pandas as pd

predictor = TabularPredictor.load("AutogluonModels/ag-20231021_122418")

while True:
    # Collect user input as a Pandas DataFrame
    user_input_data = pd.DataFrame({
        'Age': [int(input("Enter your Age in years: "))],
        'Smoker': [int(input("Are you a smoker? (0 for No, 1 for Yes): "))],
        'AlcoholConsumption': [int(input("Do you regular ingest Alcohol? (0 for No, 1 for Yes): "))],
        'HealthFood': [int(input("Do you eat healthy? (0 for No, 1 for Yes): "))],
        'PhysicalActivity': [int(input("Do you practice Physical Activity? (0 for No, 1 for Yes): "))],
        'FirstMenarcAge': [int(input("Enter your first menarc age in years: "))],
        'MenstrualCycleRegularity': [int(input("Your menstrual cycle is regular? (0 for No, 1 for Yes): "))],
        'MenstrualBleedingFlow': [int(input("Your menstrual bleeding flow is (0 for Light, 1 for Moderate, 2 for Intense): "))],
        'MenstrualPainAfterFirstPeriod': [int(input("How long it took after first menstrual period to start having menstrual pain?\n (0-Up to 1 month, 1-Between 1month and 6months, 2-Between 6months and 1year, 3-After 1year): "))],
        'PreviousPregnancy': [int(input("Have you been pregnant before? (0 for No, 1 for Yes): "))],
        'SexualActivity': [int(input("Do you have a active sex life? (0 for No, 1 for Yes): "))],
        'MenstrualPainHistoric': [int(input("Did your mother had Menstrual Pain? (0 for No, 1 for Yes): "))],
        'SiteofPain': [int(input("Is your Pain local or referred? (0-Local 'uterus region', 1-Referred'other body parts'): "))],
        'UseofMedication': [int(input("Do you use medication for menstrual pain?  (0-No, 1-Yes and Resolve Pain, 2-Yes and do not Resolve Pain): "))],     
    })

    # Make predictions using the model
    predictions = predictor.predict(user_input_data, as_pandas=True)
    PredictedMenstrualPainIntensity = predictions.iloc[0]  # Extract the predicted value

    # Add the predicted value to the user input data
    user_input_data["MenstrualPainIntensity"] = PredictedMenstrualPainIntensity

    # Append user input with predictions to the CSV file
    user_input_data.to_csv('Data\menstrualdata.csv', mode='a', header=False, index=False)

    # Display the predicted value to the user
    print(f"Your predicted Menstrual Pain Intensity: {PredictedMenstrualPainIntensity}")

    another_input = input("Do you want to input more data? (y/n): ")
    if another_input.lower() != 'y':
        break
```
