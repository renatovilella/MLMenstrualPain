## Menstrual Pain ML Predictor

### Data set variables Description

###### Age (in years)
###### Smoker (0-No, 1-Yes)
###### Alcohol Consumption (0-No, 1-Yes)
###### Health Food Intake (0-No, 1-Yes)
###### Practice of Physical Activity? (0-No, 1-Yes) 
###### First Menstrual Period 'menarc' (in years)
###### Regularity of Menstrual Period (0-No, 1-Yes)
###### Menstrual Flow 'bleeding' (0-Light, 1-Moderate, 2-Intense)
###### First Menstrual Pain after First Menarc (0-Up to 1 month, 1-Between 1month and 6months, 2-Between 6months and 1year, 3-After 1year
###### Previous Pregnancy? (0-No, 1-Yes)
###### Menstrual Pain Intensity (0-Light, 1-Moderate, 2-Intense)
###### Pain Site (0-Local 'uterus region', 1-Referred'other body parts')
###### Use of Medication when in Menstrual Pain? (0-No, 1-Yes and Resolve Pain, 2-Yes and do not Resolve Pain)

## Imports

```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
import numpy as np
```

## Data Gathering

```python
data = TabularDataset("Data\menstrualdata.csv")
menstrualdatatocomplete = TabularDataset("Data\menstrualdatatocomplete.csv")
```

```python
train_size = len(data)//3
seed = 4
train_data = data.sample(train_size, random_state=seed)
test_data = data.drop(train_data.index)
```

```python
save_path = 'menstrual_predictors_preset'
presets = ["best_quality"]
predictor = TabularPredictor(label="MenstrualPainIntensity", path=save_path)
predictor.fit(train_data, presets=presets, excluded_model_types=["NN_TORCH", "FASTAI"])
```

```python
predictor.leaderboard()
```

```python
y_test = test_data["MenstrualPainIntensity"]
test_data = test_data.drop(columns=["MenstrualPainIntensity"])

y_pred = predictor.predict(test_data)
metrics = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
save_path = 'menstrual_predictors_preset2'
presets = ["best_quality"]
predictor = TabularPredictor(label="MenstrualPainIntensity", path=save_path)
predictor.fit(train_data, presets=presets, time_limit=1800)
```

```python
y_pred = predictor.predict(test_data)
metrics = predictor.evaluate_predictions(y_true=y_test, y_pred=y_pred, auxiliary_metrics=True)
```

```python
save_path = 'menstrual_predictors_preset2'  # This was the best model from above
save_path_clone = save_path + '_clone'

predictor = TabularPredictor.load(save_path)
path_clone = predictor.clone(path=save_path_clone)
```
